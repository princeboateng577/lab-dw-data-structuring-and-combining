```python
# Your code goes here
#Read data from the urls
import pandas as pd
file1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
file2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
file3 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")

#Display the few rows of the data 
display(file1.head())
display(file2.head())
display(file3.head())

```

```python
#Display the few rows of the data 
display(file1.tail())
display(file2.tail())
display(file3.tail())
```

```python
#Display the few rows of the data 
display(file1.info())
display(file2.info())
display(file3.info())
```

```python
# Rename columns in file1
file1.rename(columns={
    "Customer": "Customer",
    "ST": "State",
    "GENDER": "Gender",
    "Education": "Education",
    "Customer Lifetime Value": "Customer Lifetime Value",
    "Income": "Income",
    "Monthly Premium Auto": "Monthly Premium Auto",
    "Number of Open Complaints": "Number of Open Complaints",
    "Policy Type": "Policy Type",
    "Vehicle Class": "Vehicle Class",
    "Total Claim Amount": "Total Claim Amount"
}, inplace=True)

# Preview to verify the changes
display(file1.head())
#file1.to_csv("cleaned_file1.csv", index=False)


```

```python
# Rename columns in file2
file2.rename(columns={
    "Customer": "Customer",
    "ST": "State",
    "GENDER": "Gender",
    "Education": "Education",
    "Customer Lifetime Value": "Customer Lifetime Value",
    "Income": "Income",
    "Monthly Premium Auto": "Monthly Premium Auto",
    "Number of Open Complaints": "Number of Open Complaints",
    "Policy Type": "Policy Type",
    "Vehicle Class": "Vehicle Class",
    "Total Claim Amount": "Total Claim Amount"
}, inplace=True)

# Preview to verify the changes
display(file2.head())

# Save the cleaned file2 to a new CSV file
#file2.to_csv("cleaned_file2.csv", index=False)

```

```python
# Reorder the columns
file2 = file2[[
    "Customer",
    "State",
    "Gender",
    "Education",
    "Customer Lifetime Value",
    "Income",
    "Monthly Premium Auto",
    "Number of Open Complaints",
    "Policy Type",
    "Vehicle Class",
    "Total Claim Amount"
]]

# Preview the reordered DataFrame
display(file2.head())
#file2.to_csv("cleaned_file2.csv", index=False)



```

```python
# Reorder the columns
file3 = file3[[
    "Customer",
    "State",
    "Gender",
    "Education",
    "Customer Lifetime Value",
    "Income",
    "Monthly Premium Auto",
    "Number of Open Complaints",
    "Policy Type",
    "Vehicle Class",
    "Total Claim Amount"
]]

# Preview the reordered DataFrame
display(file3.head())
#file3.to_csv("cleaned_file3.csv", index=False)
```

```python
# Remove non-numeric characters (like '%' or ',') and convert to float
file1['Customer Lifetime Value'] = file1['Customer Lifetime Value'] \
    .str.replace('%', '', regex=False) \
    .str.replace(',', '', regex=False) \
    .astype(float)

# Verify the conversion
print(file1['Customer Lifetime Value'].head())
print(file1['Customer Lifetime Value'].dtype)



```

```python
display(file1.info())
```

```python
# Remove the '%' sign and convert to float
file2['Customer Lifetime Value'] = file2['Customer Lifetime Value'].str.replace('%', '')
file2['Customer Lifetime Value'] = file2['Customer Lifetime Value'].astype(float)

# Preview the column to verify the conversion
print(file2['Customer Lifetime Value'].head())
```

```python
display(file1.info())
display(file2.info())
display(file3.info())
```

```python


# Function to rename columns for consistency
def rename_columns(df):
    df.rename(columns={
        "Customer": "Customer",
        "ST": "State",
        "GENDER": "Gender",
        "Education": "Education",
        "Customer Lifetime Value": "Customer Lifetime Value",
        "Income": "Income",
        "Monthly Premium Auto": "Monthly Premium Auto",
        "Number of Open Complaints": "Number of Open Complaints",
        "Policy Type": "Policy Type",
        "Vehicle Class": "Vehicle Class",
        "Total Claim Amount": "Total Claim Amount"
    }, inplace=True)
    return df

# Rename columns for file1, file2, and file3
file1 = rename_columns(file1)
file2 = rename_columns(file2)

# Reorder columns in file3 to match the others
file3 = file3[[
    "Customer",
    "State",
    "Gender",
    "Education",
    "Customer Lifetime Value",
    "Income",
    "Monthly Premium Auto",
    "Number of Open Complaints",
    "Policy Type",
    "Vehicle Class",
    "Total Claim Amount"
]]

# Combine the datasets into a single dataframe
combined_data = pd.concat([file1, file2, file3], axis=0, ignore_index=True)

# Remove duplicates
combined_data.drop_duplicates(inplace=True)

# Reset the index
combined_data.reset_index(drop=True, inplace=True)

# Save the combined cleaned dataset to a new CSV file
combined_data.to_csv("cleaned_combined_data.csv", index=False)

# Preview the final cleaned dataset
display(combined_data.head())

combined_data.info()
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
#A pivot table to show the average customer lifetime value (CLV) per gender and education level
pivot_table_clv = marketing_cutomer.pivot_table(
    values='customer_lifetime_value',  # Column with CLV data
    index='gender',                    # Gender as rows
    columns='education',               # Education as columns
    aggfunc='mean'                     # Aggregate function to calculate the average
)

# Round the values to 2 decimal points for better readability
pivot_table_clv = pivot_table_clv.round(2)

# Display the pivot table
print(pivot_table_clv)
```

*__Insights:__*
Higher Customer lifetime value for Females:
Female customers generally have a higher average customer lifetime value (CLV) than male customers across education levels.

Education Positively Impacts CLV:
Customers with advanced education levels (Doctorate and Master’s) have significantly higher CLVs, suggesting they are more profitable.

*__High-Value Segments:__*
Female customers with a Doctorate have the highest CLV.
These segments should be prioritized for tailored marketing and premium services.
Low-Value Segments:

Customers with a High School education (especially males) have the lowest CLV.
They may need targeted engagement strategies or cost-cutting measures.

*__Recommendations:__*
Focus on high-value, educated customers with loyalty programs or exclusive offers.
Improve engagement for low-CLV groups through targeted campaigns.
Use gender-specific marketing to leverage these trends.
